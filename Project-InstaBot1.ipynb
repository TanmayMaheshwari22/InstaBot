```python
post=driver.find_element_by_xpath('//div[contains(@class,"v1Nh3")]') # Locating the first post 
post.click() # Opening the first post
i=0 # Initiating i=0 as we have to like the first 30 posts only
while True: # Running an infinite loop and breaking when the i beccomes 29
    # Checking if the post is already liked or not
    # If it is already liked then priting the required message
    unlike=driver.find_element_by_xpath('html/body/div[5]/div[2]/div/article/div[3]/section[1]/span[1]/button/div/span')
    if unlike.get_attribute('outerHTML').split(' ')[2].split('=')[1]=='"Unlike"':
         print("The post is aleady been liked") 
    # If the post is not liked then locating the like button and clicking it        
    else:    
        like=driver.find_element_by_class_name('fr66n')
        time.sleep(2)
        like.click() 
        time.sleep(2)
    if i==29: #If i becomes 29 i.e first 30 posts have been liked then breaking from the loop
        break
    # After liking the post, going to the next post by locating the next post arrow symbol and clicking it    
    nextpost=driver.find_element_by_xpath('//a[contains(@class,"_65Bje")]') 
    nextpost.click()
    time.sleep(2) # Waiting for the next post to open
    i=i+1 # Incrementng i
# After breaking the loop closing the 30th post which is currently openend on the screen
close=driver.find_element_by_xpath('html/body/div[5]/div[3]/button/div')
close.click()
```

```python
post=driver.find_element_by_xpath('//div[contains(@class,"v1Nh3")]') # Locating the first post 
post.click()  # Opening the first post
i=0 # Initiating i=0 as we have to unlike the first 30 posts only
while True: # Running an infinite loop and breaking when the i beccomes 29
        # Checking if the post is already unliked or not
        # If it is already unliked then priting the required message
        like=driver.find_element_by_xpath('html/body/div[5]/div[2]/div/article/div[3]/section[1]/span[1]/button/div/span')
        if like.get_attribute('outerHTML').split(' ')[2].split('=')[1]=='"Like"':
            print('The post is already not liked')
        # If the post is liked then locating the like button and clicking it to unlike it     
        else:    
            unlike=driver.find_element_by_class_name('fr66n')
            time.sleep(2)
            unlike.click()
            time.sleep(2) 
        if i==29:
            break
        nextpost=driver.find_element_by_xpath('//a[contains(@class,"_65Bje")]')
        nextpost.click()
        time.sleep(2)  # Waiting for the next post to open
        i=i+1  # Incrementng i
# After breaking the loop closing the 30th post which is currently openend on the screen        
close=driver.find_element_by_xpath('html/body/div[5]/div[3]/button/div')
close.click()
```
